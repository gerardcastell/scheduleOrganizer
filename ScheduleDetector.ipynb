```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
IMAGE_PATH = 'example.jpeg'
```

```python
img = cv2.imread(IMAGE_PATH)
plt.imshow(img)
plt.show()
```

```python
reader = easyocr.Reader(['es'], gpu=False)
result = reader.readtext(IMAGE_PATH)
```
